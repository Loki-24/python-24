```python
import json

fp = open('emp.json', 'r')
emp_List=json.load(fp)
fp.close()

#print(emp_List)
#print(type(emp_List))  #List object

for emp_dict in emp_List:
    for key,value in emp_dict.items():
        print("{} : {}".format(key, value))
```
